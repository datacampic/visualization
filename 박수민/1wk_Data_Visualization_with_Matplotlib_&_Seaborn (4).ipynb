###**1. Matplotlibs**

```python
## Matplotlib : Library used for visualizing data into a various types of graph
x = range(0, 3)
y = range(0, 3)

# state machine interface
# plt.plot(x, y) 

# plt.title("hello")
# plt.xlabel("x")
# plt.ylabel("y")

# object oriented interface
fig, ax = plt.subplots(figsize=(3, 3))   # plt.subplots() 으로부터 fig, axis 생성
ax.plot(x, y)

ax.set_title("hello 2")
# ax.set_xticks(range(0, 3))
# ax.set_yticks(range(0, 3))
ax.set_xlabel("x")
ax.set_ylabel("y")
fig.set_dpi(300)   # dpi : dot per inch : resolution
```

####**1-0. Matplotlib Structure**
- Figure : 그래프가 들어있는 상자(전체 도화지)
- Axes : 그래프
- Grid : 그래프 내 격자
- Major tick, Minor tick : 크고 작은 눈금
- marker(line or scatter) : legend (범례) 

```python
# Axes
x = np.linspace(-np.pi*4, np.pi*4, 100)    # (start, end, dpi)
fig, axes = plt.subplots(1, 2, figsize=(8, 2))  
axes[0].plot(x, np.sin(x), linestyle=":", marker="*", color="red")   
axes[1].plot(x, np.cos(x), linestyle=":", marker="*", color="blue")    

# linestyle : "-" solid / "--" dashed / "-." dashdot / ":" dotted 
# color : "r" / green / 0.8(black(0)-1) / #524FA1
# marker : "." / "o" / "v" / "s" / "*"
```

```python
# Multiple plots in 1 ax
fig, ax = plt.subplots(figsize=(5, 2))  
x = np.linspace(-15, 15, 200) 

ax.plot(x, 3*x, label="y=x")      
ax.plot(x, x**2, label="y=x^2")

ax.legend(loc='upper right', shadow=True, fancybox=True, borderpad=1)
# fancybox : 라운드 모서리 / borderpad : 범례 박스 크기

ax.set_xlim(-5, 5)
ax.set_ylim(0, 20)
```

####**1-2. Scatter Plot**

```python
# Scatter Plot 
fig,ax = plt.subplots(figsize=(3, 3))
x = np.arange(-10, 11)
ax.plot(x, x**2,  "o", markersize=5, markerfacecolor="white", markeredgecolor="green")
```

####**1-3. Bar Plot**
- Unstacked
- Stacked

```python
# Unstacked
x = np.arange(10)
fig, ax = plt.subplots(figsize=(12,4))  # (가로, 세로)
ax.bar(x, x**2)
```

```python
# Stacked
x = [1, 2, 3]     # 1st stack
y = [1, 2, 3]     # 2nd stack
z = [1, 2, 3]     # 3rd stack
data = [x, y, z]

fig, ax = plt.subplots()   # subplot 이 1개라 위에다 쌓일 거임
x_ax = np.arange(3)     # 0, 1, 2
for i in x_ax:
    ax.bar(x_ax, data[i], bottom=np.sum(data[:i], axis=0))  # bottom : 쌓는 bar 의 시작이 어디서부터인지 표시
    ax.set_xticks(x_ax)   
    ax.set_xticklabels(["A", "B", "C"])  

plt.show()
```

```python
# Histogram : 특정 구간 내의 관측값의 도수가 막대 그래프의 높이
fig, ax = plt.subplots()
data = np.random.rand(1000)
ax.hist(data, bins=100)    # bins = 계급구간의 개수

plt.show()
```

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
```

**plt.subplots**
- specify the number of columns and rows of figure, and the overall figure size

**plt.subplot2grid**
- use the grid system to make the layout different depending on the subplot
- first, design the layout as you want
- second, set the params to visualize the figure as you intended
  - ax supposed to be the same as the number of subplots added

**plt.add_axes()**
- create an ax where you want
- can locate the subplots in a much more flexible way by specifying the location with following form, [x, y, dx, dy]
- just make sure the location of each subplot doesn't conflict each other

**gridspec, add_subplot**
- similar to subplot2grid
- but through indexing the 'grid list'
- first, create the gridspec(gs) with the size specified by tuple (row, col)
- add the subplots using gs[row, col]

**inset_axes**
- add additional axes inside of one subplot
- specify the location of inner axes with following structure 
- start from left_bottom end / [start_x, start_y, size_x, size_y]

**make_axes_locatable**
- divide or append new axes from existing axes

```python
# plt.subplots

fig, axes = plt.subplots(2, 3, figsize=(8, 5))
plt.tight_layout()
plt.show()
```

```python
# plt.subplot2grid

fig = plt.figure(figsize=(8, 5))

ax = [None for _ in range(6)]
ax[0] = plt.subplot2grid((3,4), (0,0), colspan=1)         
ax[1] = plt.subplot2grid((3,4), (0,1), colspan=1)
ax[2] = plt.subplot2grid((3,4), (0,2), colspan=1, rowspan=2)
ax[3] = plt.subplot2grid((3,4), (0,3), colspan=1, rowspan=4)
ax[4] = plt.subplot2grid((3,4), (1,0), colspan=2,rowspan=1)
ax[5] = plt.subplot2grid((3,4), (2,0), colspan=3)


for i in range(6): 
    ax[i].set_title('ax[{}]'.format(i+1)) # make ax title for distinguish:)
    ax[i].set_xticks([]) # to remove x ticks
    ax[i].set_yticks([]) # to remove y ticks
    
fig.tight_layout()
plt.show()
```

```python
# fig.add_axes()

fig = plt.figure(figsize=(8, 5))

ax = [None for _ in range(4)]
ax[0] = fig.add_axes([0.1, 0.6, 0.8, 0.4])     # [start_x, end_y, len_x, len_y]
ax[1] = fig.add_axes([0.2, 0.1, 0.5, 0.4])
ax[2] = fig.add_axes([0.8, 0.1, 0.3, 0.4])

for i in range(3):
    ax[i].set_title(f"ax[{i}]")
    ax[i].set_xticks([])
    ax[i].set_yticks([])

plt.show()
```

```python
# add_gridspec + add_subplots

fig = plt.figure(figsize=(8, 5))
gs = fig.add_gridspec(3, 4)

ax = [0 for _ in range(6)]    # size of ax = the number of the subplots you want to add

ax[0] = fig.add_subplot(gs[0, :2])
ax[1] = fig.add_subplot(gs[0, 2])
ax[2] = fig.add_subplot(gs[:, 3])
ax[3] = fig.add_subplot(gs[1, :-1])
ax[4] = fig.add_subplot(gs[2, 0])
ax[5] = fig.add_subplot(gs[2, 1:3])

for i in range(len(ax)):
    ax[i].set_title(f"ax[{i}]")
    ax[i].set_xticks([])
    ax[i].set_yticks([])
```

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

import seaborn as sns
import pandas as pd

netflix = pd.read_csv("/content/netflix_titles.csv")
netflix.head(5)
```

####**Analysis of Movies vs TV show**

```python
netflix['type'].value_counts()
netflix_tv = netflix[netflix['type'] == 'TV Show']
netflix_movie = netflix[netflix['type'] == 'Movie']

sns.set(style="darkgrid")
ax = sns.countplot(x = "type", data=netflix, palette="Set2")
```

####**The Number of Contents Added Per Year**

```python
netflix_date = netflix['date_added'].dropna()       # df 형태로 살리고 싶으면 netflix[['date_added']].dropna()

netflix_date.head(5)
netflix['year'] = netflix_date.apply(lambda x : x.split(', ')[-1])
netflix['month'] = netflix_date.apply(lambda x : x.lstrip().split(' ')[0])

# netflix.info()
```

```python
# Bar plot ordered by year

plt.figure(figsize=(8, 5))
sns.set(style="darkgrid")
year_order = list(map(str, range(2008, 2021)))
ax = sns.countplot(x="year", data=netflix, palette="Set2", order=year_order)    # netflix['year'].value_counts().index

plt.show()
```

```python
# Heat Map (Month, Year)

month_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
df = netflix.groupby('year')['month'].value_counts().unstack().fillna(0)[month_order]

plt.figure(figsize=(8, 5), dpi=200)
plt.pcolor(df, cmap='Blues', edgecolors='white', linewidths=2)     # heatmap
plt.xticks(np.arange(0.5, len(df.columns), 1), df.columns, rotation='vertical', fontsize=7, fontweight='bold')      # column : month
plt.yticks(np.arange(0.5, len(df.index), 1), df.index, fontsize=7)          # index : year
plt.title('Netflix Contents Update', fontsize=12, fontweight='bold', position=(0.20, 1.0+0.02))

cbar = plt.colorbar()
cbar.ax.tick_params(labelsize=6) 
# cbar.ax.minorticks_on()

plt.show()
```

####**Top 10 Contents Creating Countries**

```python
# Bar Plot (Horizontal)

plt.figure(figsize=(15, 15))
sns.set(style="darkgrid")

netflix['country'].fillna('Unknown', inplace=True)
ax = sns.countplot(y="country", data=netflix, palette="Set2", 
                   order=netflix['country'].value_counts().index[:50])    # netflix['year'].value_counts().index

plt.yticks(rotation="horizontal")
plt.show()
```

```python
# Pie Chart (WedgeProps)
country = netflix['country'].value_counts(ascending=False)[:10]

explode = [0.02 for _ in range(len(country))]
wedgeprops={'width': 0.7, 'edgecolor': 'w', 'linewidth': 5}   
# width : in a ratio to radius

figure = plt.figure(figsize=(12, 15))
plt.pie(country, labels=country.index, autopct='%.1f%%', startangle=80, 
        explode=explode, wedgeprops=wedgeprops, counterclock=False)    # shadow=True
plt.title("Top 10 Contents Creating Countries", fontsize=20, fontweight='bold')
plt.legend(loc='upper right', shadow=True, fancybox=True, borderpad=0.1)

plt.show()
```

####**Distribution of Movie Rating**

```python
netflix_movie = netflix[netflix['type']== 'Movie']

plt.figure(figsize=(10, 8))
sns.set(style="darkgrid")
rating_order =  ['G', 'TV-Y', 'TV-G', 'PG', 'TV-Y7', 'TV-Y7-FV', 'TV-PG', 'PG-13', 'TV-14', 'R', 'NC-17', 'TV-MA']
ax = sns.countplot(x="rating", data=netflix_movie, palette="Set2", 
                   order=rating_order)    # netflix['year'].value_counts().index

plt.xticks(rotation="vertical")
plt.show()
```

```python
# original code : https://www.kaggle.com/subinium/storytelling-with-data-netflix-ver
sns.set_style('white')

def rating_barplot(data, title, height, h_lim):
    fig, ax = plt.subplots(1,1, figsize=(14, 7), dpi=200, facecolor='white')
    ax.set_ylim(0, h_lim)
    ax.bar(data.index, data,  color="w", width=0.52, edgecolor='black')

    color =  ['green',  'blue',  'orange',  'red']
    span_range = [[0, 2], [3,  6], [7, 8], [9, 11]]

    for i, rank in enumerate(['Little Kids', 'Older Kids', 'Teens', 'Mature']):
        ax.annotate(rank,
                    xy=(sum(span_range[i])/2 ,height),
                    xytext=(0,0), textcoords='offset points',
                    va="center", ha="center",
                    color="w", fontsize=16, fontweight='bold',
                    bbox=dict(boxstyle='round4', pad=0.4, color=color[i], alpha=0.6))
        if i == 0:
            ax.axvspan(-0.5, 2.5,  color=color[0], alpha=0.07)
        elif i == 3:
            ax.axvspan(8.55, 11.5,  color=color[3], alpha=0.07)
        else:
          ax.axvspan(span_range[i][0]-0.45, span_range[i][1]+0.5,  color=color[i], alpha=0.07)
    
    
    ax.set_title(f'Distribution of {title} Rating', fontsize=15, fontweight='bold', position=(0.48, 1.0+0.05))
    plt.show()
    

rating_order =  ['G', 'TV-Y', 'TV-G', 'PG', 'TV-Y7', 'TV-Y7-FV', 'TV-PG', 'PG-13', 'TV-14', 'R', 'NC-17', 'TV-MA']
movie_rating = netflix_movie['rating'].value_counts()[rating_order]

rating_barplot(movie_rating,'Movie', 1600, 2000)
```

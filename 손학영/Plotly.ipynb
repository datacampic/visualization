# 03. 총지출

```python
df_out = pd.read_excel("/content/drive/MyDrive/스터디/data/1. 예산/1-2-2. 세출 지출 세목 예산편성현황(총지출).xlsx")
df_out.head()
```

```python
df_out = df_out[df_out['소관명']=='문화체육관광부']
df_out.shape
```

```python
df_out.head()
```

```python
for i in df_out.columns[:-3] : 
  print(i, ':', df_out[i].unique())
  print()
```

```python
df_out.drop(['소관명', '분야명'], axis = 1 , inplace = True)
```

```python
# 일단은 부문명 별로 확인해보기 
df_sum = pd.crosstab(index = df_out['부문명'], columns = df_out['회계연도'], values =df_out['국회확정금액(천원)'], aggfunc = np.sum)
df_sum.reset_index(inplace=True)
df_sum
```

```python
df_sum = df_sum.T
df_sum = df_sum.rename(columns=df_sum.iloc[0])
df_sum.drop(df_sum.index[0], inplace = True)
df_sum.reset_index(inplace = True)
df_sum
```

```python
# 연도별 부문에 따른 지출 금액 변화 확인 

fig = go.Figure() # 그래프 그릴 fig 생성 

for i in df_sum.columns[1:] : 
  fig.add_trace(go.Scatter(x = df_sum['회계연도'], 
                            y = df_sum[i]/10000000,
                           name = i))

fig.update_layout(title = '연도별 부문에 따른 지출 금액 변화',
                  xaxis_title = '연',
                  yaxis_title = '국회확정금액(백억원)',
                  plot_bgcolor = "#f4f4f4")
fig.update_xaxes(
    type='category'
)
fig.show()
```

문화예술 부문은 코로나 이후로도 계속 증가했네?! 심지어 **더 가파르게 증가함**!
